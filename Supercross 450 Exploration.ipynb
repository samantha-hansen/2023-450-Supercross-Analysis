# Data Preparation

```python
#Replace incorrect date
```

```python
df['Race_Date'] = df['Race_Date'].replace('2/32023','3/2/2023')
```

```python
#Alter data type of selected columns
```

```python
df['Race_Date'] = pd.to_datetime(df['Race_Date'])
```

```python
df.dtypes
```

```python
#Replace values in 'Holeshot' column
```

```python
df['Holeshot'] = df['Holeshot'].replace('X', 'True')
df['Holeshot'].fillna(False, inplace=True)
```

```python
df.head(20)
```

```python
#Alter datatype of 'Holeshot' column
```

```python
df['Holeshot'] = df['Holeshot'].astype('bool')
```

```python
df.dtypes
```

```python
#Check for missing values in data
df.isna().sum()
```

```python
#Missing values explained as follows:
    #Race_Date,Race_Name,Race_State-only listed once for each race rather than in every line corresponding to that race
    #First_Lap_Standing and Laps_Led data not tracked by AMA for triple crown format races
    #Triple Crown values missing in regular race format placing columns
```

# Feature Understanding

```python
#Breakdown of number of each bike make in season-wide main events
df['Bike_Make'].value_counts()
```

```python
ax = df['Bike_Make'].value_counts() \
    .head(9) \
    .plot(kind='bar', title = 'Top Bike Make')
ax.set_xlabel('Bike Make')
ax.set_ylabel('Count')
plt.show
```

```python
#Examine rider AMA points
```

```python
rider_points = df.groupby('Rider_Name')['Points'].sum()
rider_points = rider_points.sort_values(ascending=False)
print(rider_points)
```

```python
ax = rider_points \
    .head(15) \
    .plot(kind='bar', title = 'Top 15 450 Class Riders')
ax.set_xlabel('Rider')
ax.set_ylabel('Points')
plt.show
```

```python
#Examine rider PulpMX fantasy points
```

```python
fantasy_points = df.groupby('Rider_Name')['Fantasy_Points'].sum()
fantasy_points = fantasy_points.sort_values(ascending=False)
print(fantasy_points)
```

```python
ax = fantasy_points \
    .head(20) \
    .plot(kind='bar', title = 'Top 20 450 Class Fantasy Riders')

ax.set_xlabel('Rider')
ax.set_ylabel('Points')
plt.xticks(rotation=60, ha='right')
plt.show
```

```python
#Examine the breakdown of rider homestates
```

```python
#Extract home state from 'Hometown' column
df['Homestate'] = df['Hometown'].str.split(',').str[-1].str.strip() 

#create new df w unique rider names and home states
unique_riders = df.drop_duplicates(subset=['Rider_Name'])
home_state = unique_riders['Homestate'].value_counts().reset_index()
home_state.columns = ['Homestate', 'Count']

#create chart
homestate = px.choropleth(home_state,
                         locations='Homestate',
                         locationmode="USA-states",
                         scope="usa",
                         color='Count',
                         title='Rider Home States',
                         color_continuous_scale='portland',
                         width=600,
                         height=400
                         )
homestate.update_layout(
    title_x=0.5)
homestate.show()
```

```python
#Examine the breakdown of races by state
```

```python
state_counts = df['Race_State'].value_counts()
state_data =pd.DataFrame({'State': state_counts.index, 'Count':state_counts.values})

#Create chart
states = px.choropleth(state_data,
                      locations='State',
                      locationmode="USA-states",
                      scope="usa",
                      color='Count',
                      color_continuous_scale="portland",
                       width=600,
                       height=400
                      )
states.update_layout(
    title_text = 'Number of Races per State',
    title_x = 0.5)
states.show()
```

# Feature Relationships

```python
qualfin = px.scatter(df, x='Qualifying', 
                     y='Finish', 
                     color='Fantasy_Points', 
                     hover_data=['Rider_Name', 'Fantasy_Handicap'])
qualfin.update_layout(title='Qualifying Rank vs. Official Finish',
                     title_x= 0.5,
                     width=600,
                     height=400)
qualfin.show()
```

```python
#Create comparision matrix with numerical values colored by bike make
sns.set(style="ticks")
plt.figure(figsize=(12,10))

sns.pairplot(df, vars=['Finish', 'Qualifying', 'Points','Fantasy_Points', 'Fantasy_Handicap'],
            hue='Bike_Make')
plt.show()
```

```python
#Create correlation matrix to observe correlations bt features
df_corr = df[['Finish', 'Qualifying', 'Points',
               'Fantasy_Points', 'Fantasy_Handicap']].corr()
df_corr
```
